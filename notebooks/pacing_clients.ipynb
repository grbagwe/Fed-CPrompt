```python
import numpy as np
```

```python
class Args():
    pass
args = Args()
```

```python
args.num_clients = 10
args.num_rounds = 20
args.num_tasks = 10
L_subset_clients = np.arange(0, args.num_clients)
group_1 = {0,1,2,3,4}
group_2 = {5,6,7,8,9}
```

```python
client_task = {}
for client in range(args.num_clients): # init task to clients
    client_task[client] = 0

for task_no in range(args.num_tasks):
    for round_ in range(args.num_rounds):
        for client in L_subset_clients:
            # if client in group_1:
            if round_ == int(args.num_rounds/2 ):
                if client in group_2:
                    client_task[client]  = task_no + 1
                    print('original task', task_no, 'task', client_task[client] , 'client', client, 'round', round_)
                else:
                    print('original task', task_no, 'task', client_task[client] , 'client', client, 'round', round_)
            else :
                if client in group_2:
                    # client_task[client]  = task_no + 1
                    print('original task', task_no, 'task', client_task[client] , 'client', client, 'round', round_)
                else:
                    print('original task', task_no, 'task', client_task[client] , 'client', client, 'round', round_)

    for client in range(args.num_clients): # init task to clients
        client_task[client] = task_no + 1

```
